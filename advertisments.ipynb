```python
##Machine Learning model-Logistic Regression
```

```python
##Spliting the data into training set and testing set using train_test_split
```

```python
from sklearn.model_selection import train_test_split

X = ad_data.drop(['Clicked on Ad','Timestamp','City','Country','Ad Topic Line'], axis =1)
y = ad_data['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
## Training and fitting a logistic regression model on the training set
```

```python
ad_data.columns
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmod = LogisticRegression()
```

```python
logmod.fit(X_train, y_train)
```

```python
##Predictions and Evaluations
```

```python
predictions = logmod.predict(X_test)
```

```python
##Creating a classification report
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
